```python
import re

f= open("abs1.txt","r+")

data = f.read()
split_file_data=(re.sub("\n","\n\n",data))
split_file = open("abs1.split.txt","w+")
split_file.write(split_file_data)
split_file.close()


f.close()

```

```python
# 3. Finds number of mentions of 'customers', 'investors', 'stakeholders'
# (accounting for different ways of expressing these concepts) in each (sub)document.
```

```python
import re
f= open("abs1.txt","r+")
line = f.readline()
i = 0
while line:
    para = line.split()
    wordCount = len(para)
    customers    = len(re.findall(r'[cC]ustomers+',line))
    investors    = len(re.findall(r'[iI]nvestors+',line))
    stakeholders = len(re.findall(r'[sS]takeholders+',line))

    print ("Sub Document "+ str(i+1)+" :customers : " + str(customers)  + ", investors: " + str(investors) + ", stakeholders: " + str(stakeholders) + "");

    i = i+1
    line = f.readline()
f.close()



```
